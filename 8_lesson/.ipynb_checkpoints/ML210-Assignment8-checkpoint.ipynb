# Assignment 8: Moving Beyond Linearity

Dataset(s) needed:
Hitters data frame of the ISLR library SF-DAT-21. In this assignment, we're going to predict Salary based on Hits as the only independent variable.

```python
import pandas as pd
import numpy as np

train = pd.read_csv('Hitters.csv')
train.head(10)
```

<h3> Q.1. Replace missing values in the dataset by the mean</h3>

```python
train.isnull().sum()
train.mean()
```

```python
train = train.fillna(train.mean())
```

<h3> Q.2. Graph a scatter plot to visualize the relationship between Hits and Salary</h3>

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(x = train['Salary'], y = train['Hits'])
plt.xlabel('Hits')
plt.ylabel('Salary')
```

<h3> Q.3. Apply Linear Regression on the Hitters data to predict 'Salary' depending only on 'Hits' as the independent variable. Print out the slope and intercept coefficients.
</h3>

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# TODO: Split the data into 70/30 train/test datasets
X = train[['Hits']]
y = train[['Salary']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=42)

# TODO: Fit linear regression model
reg = LinearRegression().fit(X_train, y_train)

# TODO: print the slope and intercept
print("Linear Regression Intercept: ",reg.intercept_[0])
print("Linear Regression Coef: Salary -",reg.coef_[0])
```

<h3> Q.4. Compute the RMSE of the model obtained above</h3>

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

lm_pred = reg.predict(X_test)

print('Linear Reg Root Mean Squared Error:\n', sqrt(mean_squared_error(y_test, lm_pred)))
```

<h3> Q.5. Graph the linear regression model obtained on the data</h3>

```python
linear_regr = np.linspace(X_test['Hits'].min(), X_test['Hits'].max())
linear_regr = linear_regr.reshape(-1,1)
plot = reg.predict(linear_regr)

plt.scatter(X_test['Hits'], y_test['Salary'], facecolor='None', edgecolor='k', alpha=0.7)
plt.plot(linear_regr, plot, label='Linear Regression')
plt.legend()
plt.ylabel("Salary")
plt.xlabel("Hits")
plt.show()
```

<h3> Q.6. Construct a polynomial regression model with degree 2 to predict 'Salary' depending only on 'Hits' as the independent variable. Compute the RMSE and Plot your result.
</h3>

```python
import numpy as np

# Generate weights for polynomial function with degree 2
weights = np.polyfit(X_train['Hits'], y_train['Salary'], 2)
# print(weights)

# Generate model with the given weights
poly_model = np.poly1d(weights)

# Prediction on test set
poly_pred = poly_model(X_test['Hits'])

# TODO: Compute the RMSE
print('Poly Reg Root Mean Squared Error:\n', sqrt(mean_squared_error(y_test, poly_pred)))
# TODO: Plot the visualisation
poly_regr = np.linspace(X_test.min(),X_test.max())
plot = poly_model(poly_regr)

plt.scatter(X_test, y_test, facecolor='None', edgecolor='k', alpha=0.3)
plt.plot(poly_regr, plot, label='Polynomial Regression')
plt.legend()
plt.xlabel('Hits')
plt.ylabel('Salary')
plt.show()
```

<h3> Q.7. Create a cubic spline with 3 knots at 40, 100 and 150</h3>

```python
from patsy import dmatrix
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Generate a cubic spline with 3 knots at 25, 40 and 60
transformed_x = dmatrix("bs(train, knots=(40, 100, 150), degree=3, include_intercept=False)",
                        {"train": X_train['Hits']},return_type='dataframe')

# Fit a generalized linear models on the transformed dataset
fit_glm = sm.GLM(y_train['Salary'], transformed_x).fit()

# Prediction
pred_glm = fit_glm.predict(dmatrix("bs(valid, knots=(40, 100, 150), include_intercept=False)",
                                   {"valid": X_test['Hits']}, return_type='dataframe'))

# TODO: Compute the RMSE 
print('Cubic spline Reg Root Mean Squared Error:\n', sqrt(mean_squared_error(y_test, pred_glm)))
# TODO: Plot the visualisation
lin = np.linspace(X_test['Hits'].min(), X_test['Hits'].max(), 97)
pred_glm = fit_glm.predict(dmatrix("bs(xp, knots=(40, 100, 150), degree=3, include_intercept=False)",
                                   {"xp": lin}, return_type='dataframe'))

plt.scatter(X_test['Hits'], y_test['Salary'], facecolor='None', edgecolor='k', alpha=.7)
plt.plot(lin, pred_glm, label = 'Cubic Spline with 3 knots')
plt.legend()
plt.xlabel('Hits')
plt.ylabel('Salary')
plt.show()
```

<h3> Q.8. Create a natural cubic spline on the data to predict 'Salary' based on 'Hits'</h3>

```python
# Generate natural cubic spline
transformed_x = dmatrix("cr(train, df=3)",
                        {"train": X_train['Hits']},return_type='dataframe')
fit_spline = sm.GLM(y_train['Salary'], transformed_x).fit()

# TODO: Compute the RMSE 
pred_nat = fit_spline.predict(dmatrix("cr(valid, df=3)", {"valid": X_test['Hits']}, return_type='dataframe'))
print('Natural spline Reg Root Mean Squared Error:\n', sqrt(mean_squared_error(y_test, pred_nat)))
# Visualization

xp = np.linspace(X_test['Hits'].min(), X_test['Hits'].max(), 97)
pred_nat = fit_spline.predict(dmatrix("cr(xp, df=3)", {"xp": xp}, return_type='dataframe'))

# TODO: Plot the visualisation
plt.scatter(X_test['Hits'], y_test['Salary'], facecolor='None', edgecolor='k', alpha=.7)
plt.plot(xp, pred_nat, color='r', label='Natural spline')
plt.legend()
plt.xlabel('Hits')
plt.ylabel('Salary')
plt.show()
```

<h3> Q.9. Which model performed the best?</h3>

The polynomial regression with degree 2 returned the best RMSE followed by the natural spline model. The linear regression seems to have under fit the data and the cubic spline with 3 knots seems to have overfit the training data.
