<p>In this exercise you will use the US Adult Census data relating income to social factors such as Age, Education, race etc...</p>

<p>The goal here is to train a binary classifier on the training dataset to predict the column income_bracket which has two possible values ">50K" and "<=50K" and evaluate the accuracy of the classifier with the test dataset.</p>


```python
import pandas as pd
import numpy as np
from sklearn import preprocessing, metrics
from sklearn.model_selection import train_test_split


train=pd.read_csv('census-training.csv')
print(train.info())
train.head(20)
# train.shape
```

<h3>Q.1. Write a command that will calculate the number of unique values for each feature in the training data.</h3>

```python
#TODO
train.nunique()
```

<h3>Q.2. There is a special character '?' inserted in the data in place of null values.  Write a command that will remove it and replace with NaN.</h3>

```python
# Command (s): remove the '?' special character and replace with NaN
#TODO
train = train.replace({'?': np.nan})

train.isnull().sum() #checking the dataset for NaN values .... NaN values in two columns
```

```python
train.head(20)
```

<h3>Q.3. Replace all missing values in Occupation and Country by their mode (most occuring values). </h3>

```python
#Command(s):
train.mode() # finding mode to compute the most occuring item in every column
#replace missing values with their mode

train['Occupation'].fillna(train['Occupation'].mode()[0], inplace=True)
train['Country'].fillna(train['Country'].mode()[0], inplace=True)

train.isnull().sum() #checking the data for NaN values .... should all show 0 now
```

<h3>Q.4. Write functions to replace Gender and Income attributes to "0" and "1". </h3>

```python
# converting Gender to "0" and "1" 
def gender_to_numeric(x):
    if x == 'Male':
        return 1
    else:
        return 0
    #TODO return 1 f gender is Male, 0 otherwise
    
train['Gender'] = train['Gender'].apply(gender_to_numeric)

# converting Inccome to "0" and "1" 
def income_to_numeric(x):
    if x == '<=50K.':
        return 0
    else:
        return 1
    #TODO return 1 if income is greater than 50k, 0 otherwise
    
train['Income'] = train['Income'].apply(income_to_numeric)

train.head(10)
```

<h3>Q.5. Use the label encoder API to encode all object-type in the data. </h3>

```python
# Command(s) 
obj = train.select_dtypes(include=['object']) #all features that are 'object' datatypes
le = preprocessing.LabelEncoder()
for i in range(len(obj.columns)):
    train[obj.columns[i]] = le.fit_transform(train[obj.columns[i]])

# Would it be better to leave one out one-hot this data rather than use the LabelEncoder? -
# It suggestes that the distance between 'Married' and 'Divorced' is the same as
# 'Married' and 'Never-married' and the same goes for the other object features. 
    
#Encode input data

train.head(10)
```

<h3>Q.6. Split the data into 70% training and 30% test </h3>

```python
#TODO 
y = train['Income']
X = train.drop('Income', axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=83)
```

<h3>Q.7. Train the logistic regression model on the train dataset and evaluate its performance on the test dataset.</p>

```python
#Initialize classifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, roc_auc_score, accuracy_score


log_rgr = LogisticRegression(random_state=83)
log_rgr = log_rgr.fit(X_train, y_train)
# TODO: fit X_train and y_train

#predict on the test set
y_pred = log_rgr.predict(X_test) 

lr_acc = accuracy_score(y_test, y_pred)
# calculate the fpr and tpr for all thresholds of the classification
probs = log_rgr.predict_proba(X_test)
preds = probs[:,1]
fpr, tpr, threshold = metrics.roc_curve(y_test, preds)
lr_roc_auc = roc_auc_score(y_test, y_pred)
lr_classification_report = classification_report(y_test, y_pred)
lr_confusion_matrix = confusion_matrix(y_test, y_pred)

print("Accuracy: {}".format(lr_acc))
print("AUC: {}".format(lr_roc_auc))
print("\n Confusion Matrix: \n", lr_confusion_matrix)
print("\n Classification Report: \n", lr_classification_report)

# Plot the ROC curve
import matplotlib.pyplot as plt
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % lr_roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

I think that one-hot encoding the data rather than the LabelEncoding might have helped improve the model results. 

There is a problems with class imbalance that the RFE isn't fixing. Below I re-ran the Logistic Regression with a SMOTE-Tomek and moved the ROC-AUC up to .56 while taking a rather significant hit to the Accuracy (.58) so that is a trade off that would have to be considered if finding the minority cases were important. There are also several highly collinear variables based on their VIF values. The RFE removed three of the 5 highly collinear features which might explain why there wasn't a big improvement in predictive ability.
